```python
from sklearn.datasets import load_iris 
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

```python
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame( iris.data , columns=iris.feature_names )
df
```

```python
df['target']  = iris.target
```

```python
df['flower_name']  = df['target'].apply( lambda X : iris.target_names[X] )
df
```

```python
df0 = df[:50 ]
df1 = df[50: 100 ]
df2 = df[100 : ]
```

```python
plt.xlabel( ' sepal length')
plt.xlabel( ' sepal width')
plt.scatter( df0['sepal length (cm)'] , df0['sepal width (cm)']  , marker='+' , color ='red')
plt.scatter( df1['sepal length (cm)'] , df1['sepal width (cm)']  , marker='.' , color ='green')
plt.scatter( df2['sepal length (cm)'] , df2['sepal width (cm)']  , marker='*' , color ='blue')

plt.show( )
```

```python
from sklearn.svm import SVC 
model = SVC()
X = df.drop( columns=[ 'flower_name'  , 'target'])
Y = df['target']
```

```python
from sklearn.model_selection import train_test_split 
xtr , xte , ytr , yte = train_test_split( X ,Y  ,  test_size= 0.3 )
```

```python
model.fit( xtr , ytr )
```

```python
model.score( xte , yte )
```

```python
pred = model.predict( xte )

from sklearn.metrics import accuracy_score 
score = accuracy_score( yte , pred )
print( score )
```

```python
hir['experience']  = hir['experience'].fillna( 0 )
hir 
```

```python
from word2number import w2n 
```

```python

def convertnum( value) :
    if isinstance( value, str )  and value.isalpha() : 
        return w2n.word_to_num(value)
    return value     
    

hir['experience'] = hir['experience'].apply( convertnum )
```

```python
hir 
```

```python

kaka = hir['test_score(out of 10)'].median()
hir['test_score(out of 10)'] = hir['test_score(out of 10)'].fillna(kaka)
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit( hir[['experience' ,'test_score(out of 10)' , 'interview_score(out of 10)' ]] , hir['salary($)']) 
```

```python
lm.predict( [[2,9,6]])
```

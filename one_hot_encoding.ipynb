```python
kaka = pd.read_csv("E:\DATA ANLYST\material\carprices.csv")
```

```python
kaka
```

```python
kaka['Car Model']   =  le.fit_transform( kaka['Car Model'])

kaka
```

```python
cct = ColumnTransformer( [ ( 'Car Model'  , OneHotEncoder()  ,  [0])]  , remainder='passthrough') 
```

```python
cct
```

```python
kakaji = kaka 
y = kakaji['Sell Price($)']
X= kakaji[['Car Model' ,'Mileage' ,'Age(yrs)'] ] 
X
```

```python
X = cct.fit_transform(X) 
X
```

```python
print(cct.named_transformers_)
```

```python
#X_transformed = cct.fit_transform(df)

ohe = cct.named_transformers_['Car Model']
car_feature_names = ohe.get_feature_names_out(input_features=['Car Model'])

# Combine with Remaining Column Names
new_feature_names = list(car_feature_names) + ['Mileage' ,'Age(yrs)']
```

```python
new_feature_names
```

```python
X = pd.DataFrame(  X , columns= new_feature_names)
```

```python
model
```

```python
model.fit( X , y  )
```

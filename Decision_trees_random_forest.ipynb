```python
ice = pd.read_csv('E:\\DATA ANLYST\\material\\titanic.csv')
```

```python
ice 
```

```python
''' 
Pclass
Sex
Age
Fare
'''

features  = ice [['Pclass' , 
'Sex', 
'Age' , 'Fare']   ]
```

```python
lb_mex = LabelEncoder() 

features['Sex_1']  = lb_mex.fit_transform( features['Sex'])
features
```

```python
features = features.drop( [ 'Sex_1'] , axis =1 )
```

```python
features 
```

```python
survival = ice['Survived']
survival
```

```python
ice_dtree = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split 
xtrn , xtst , ytrn , ytst =  train_test_split( features , survival , test_size= 0.3 )
```

```python
ice_dtree.fit( xtrn , ytrn )
```

```python
ice_dtree.score( xtst , ytst )
```

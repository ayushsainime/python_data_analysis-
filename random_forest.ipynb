```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import load_digits
digits= load_digits( )

```

```python
dir( digits)
```

```python
df = pd.DataFrame( digits.data)
```

```python
df['targets'] = digits.target
```

```python
df
```

```python
imagedat = digits.images[:12] 
imgtarget = digits.target[: 12 ]

plt.figure( figsize=( 12, 4 ))


for i in range( 10 ) :
    plt.subplot(2 , 7 , i+1  )
    plt.imshow( imagedat[ i]  , cmap = 'viridis') 
    plt.title( str( imgtarget[i]))
    plt.axis( 'off')

#plt.tight_layout() 
plt.show()

```

```python
X = df.drop('targets',axis='columns')
y = df.targets
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)

y_predicted = model.predict( X_test) 
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix( y_test ,y_predicted )
cm 
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
plt.Figure( figsize= ( 10 , 2 ))

sns.heatmap( cm , annot = True )
plt.xlabel( 'predicted ')
plt.ylabel( 'actual ')
```

**NOW THE EXCERSISE  USING  IRIS DATASET**

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir( iris)
```

```python
cliff = pd.DataFrame( iris.data , columns = iris.feature_names )
```

```python
cliff['target '] = iris.target
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

xtrn , xtst , ytrn , ytst = train_test_split( cliff.drop( 'target ' , axis =1 ) , cliff['target '], test_size= 0.3 )
rfc = RandomForestClassifier( n_estimators= 10 )
rfc.fit( xtrn,ytrn)
```

```python
rfc.score(xtst , ytst  )
```

```python
rfc = RandomForestClassifier( n_estimators= 40 )
rfc.fit( xtrn,ytrn)
```

```python
rfc.score(xtst , ytst  )
```

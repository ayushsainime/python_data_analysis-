```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import  seaborn as sns 
```

```python
df = pd.read_csv("E:\DATA ANLYST\material\spam.csv")
```

```python
df
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply( lambda X :  1 if X== 'spam'  else 0  ) 
df
```

```python
df.head()
```

```python
from sklearn.model_selection  import train_test_split
X_train , X_test , y_train, y_test = train_test_split( df['Message']  , df['spam'] , test_size= 0.25 )   
```

```python
from sklearn.feature_extraction.text import CountVectorizer 

v = CountVectorizer() 

X_train_count = v.fit_transform( X_train.values)
```

```python
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test) 
model.score( X_test_count ,  y_test)
```

Sklearn Pipeline




```python
from sklearn.pipeline import Pipeline 
clf = Pipeline( 
    [ 
        ( 'vectorizer'   , CountVectorizer() ) ,
        ( 'nb'  , MultinomialNB() ) 
    ]
)

clf.fit( X_train  ,y_train  )



```

```python
clf.score( X_test , y_test )
```

```python
clf.predict( emails )
```

```python
from sklearn import datasets
wine = datasets.load_wine()
```

```python
wine.feature_names 
```

```python
ww = pd.DataFrame( wine.data , columns=wine.feature_names )
```

```python
ww
```

```python
ww['target'] = wine.target
```

```python
from sklearn.model_selection import train_test_split 

wx_train , wx_test , wy_train , wy_test  = train_test_split( wine.data  , wine.target  , test_size=  0.3  , random_state= 27)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB 
from sklearn.pipeline import Pipeline 

kaka = Pipeline( 
    [
        ( 'gaussian data '  , GaussianNB())
    ]
)

kaka.fit( wx_train , wy_train)
```

```python
bagge = Pipeline(
    [
        ( ' multinomail '  , MultinomialNB())
    ]
 )

bagge.fit(wx_train , wy_train)
```

```python
bagge.score( wx_test , wy_test) 
```

```python
kaka.score( wx_test , wy_test )
```

```python
cliff = kaka.predict( wx_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score( wy_test , cliff  )  
```

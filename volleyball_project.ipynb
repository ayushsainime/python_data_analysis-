```python
attack_by_country = df.groupby('Country')["Attack"].mean()
attack_by_country.sort_values(ascending = False  , inplace =True  )
attack_by_country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure previous conflicts are removed
plt.figure(figsize=(8, 8))

sns.barplot(x=attack_by_country.index, 
            y=attack_by_country.values, 
           
            legend=False, 
            palette='viridis')

# Correct labels and title
plt.xlabel("Countries")
plt.ylabel("Attack Rate")
plt.title("Attack Average by Country")
plt.xticks(rotation=45)

plt.show()
```

```python
top_10_countries = attack_by_country.head( 10 )

plt.figure( figsize =( 8 ,6 ))
sns.barplot( x = top_10_countries.index  , y =  top_10_countries.values )
plt.xlabel(  "countries ")
plt.ylabel(" attack rate ")
plt.title( "attack average by country  top 10 ")
```
